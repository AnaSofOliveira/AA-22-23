```python
import numpy as np
import pickle
import mglearn
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay 

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
```

```python
fN = '/Users/anaso/Documents/AA/AA-22-23/Trabalho Final/imdbFull.p'

D = pickle.load(open(fN, 'rb'))

nDocs = 5000 

corpus = D.data[:nDocs]
y = D.target[:nDocs]

y = [1 if val>5 else 0 for val in y]
```

```python
print("Tamanho do corpus: {}".format(len(corpus)))
```

```python
tf = TfidfVectorizer().fit(corpus)
corpus_p = tf.transform(corpus)

features = np.array(tf.get_feature_names_out())
```

```python
print(corpus_p.shape)
```

```python
st = StandardScaler(with_mean=False).fit(corpus_p)
corpus_s = st.transform(corpus_p)

lr = LogisticRegression(max_iter=5000, C=10, penalty='l2', solver='liblinear', tol=1e-3).fit(corpus_s, y)
y_e = lr.predict(corpus_s)

w = lr.coef_
```

```python
print(w.shape)
print(features.shape)
```

```python
print(features[:20])
print(w[0, :20])
```

```python
mglearn.tools. visualize_coefficients(w, features, n_top_features=40)
```

```python
features = features[w[0]>0]
print(features[:20])
```

```python
print(features.shape)
print(type(corpus), type(features))
```

```python
tf = TfidfVectorizer().fit(features)
corpus_p = tf.transform(corpus)

st = StandardScaler(with_mean=False).fit(corpus_p)
corpus_s = st.transform(corpus_p)

new_features = tf.get_feature_names_out()
print(len(new_features))
```

```python
print(corpus_p.shape)
```

```python
lr = lr.fit(corpus_s, y)
y2_e = lr.predict(corpus_s)
```

## Comparao

```python
print("Nmero de Erros: {}".format(np.sum(y != y_e)))
print("Matriz de Confuso: \n{}".format(confusion_matrix(y, y_e)))
```

```python
print("Nmero de Erros: {}".format(np.sum(y != y2_e)))
print("Matriz de Confuso: \n{}".format(confusion_matrix(y, y2_e)))
```
